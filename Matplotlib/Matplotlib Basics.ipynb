# Line charts üìä
##  1. Line charts are typically used to visualize quantitative data(Numerical data), particularly data that varies continuously over a range.
##  2. Can be used on ordered categorical data: 
    - Line charts can be used when the categorical data has a meaningful order or sequence. Examples include:
    - Time-Series: Time (categorical) vs. Temperature (numerical) over days, months, or years.
    - Ordinal Data: Categories like low, medium, high (categorical) vs. numerical measurements or scores.

 ## Simple plots üìà

```python
# plotting a 
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
```

This graph shows the trend of prices over the years from 2015 to 2020.

- The x-axis represents the years.
- The y-axis represents the prices.
- The plot indicates that prices started around 48,000 in 2015, peaked in 2017 at 57,000, and then gradually declined to 45,000 by 2020.

### Show plot üìà
In some environments like Jupyter Notebooks or certain interactive Python environments, you can display plots without explicitly using plt.show() 

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('My Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.grid()
plt.show() #Used in other IDE just to show the graph on console
```

# Scatter Plots üìä
## A scatter plot is used to visualize the relationship between two numerical variables. Each point on the plot represents a single observation.
## Scenarios where scatter plots are used:
### 1. Correlation: Scatter plots can show the relationship and strength of correlation between two numerical variables.
### 2. Distribution: They can display the distribution of data points and help identify clusters or patterns within the data.
### 3. Outliers: Scatter plots are useful for identifying outliers or unusual observations
### 3. Comparisons: They are effective for comparing two numerical variables across different categories or groups.

- Basically Numerical vs Numerical

## Simple plots üìà

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)

y = 10*x + 3 + np.random.randint(0,300,50) # 0 to 300 random nuumber generating 50 times making an array
print(x)
print(y)
plt.scatter(x,y)
```

- This example illustrates the use of scatter plots to visualize the relationship between two variables, where one variable (y) is influenced by another (x) along with some added randomness.
- Linear Relationship with Noise(Disturbance): The scatter plot will show points that roughly follow a linear trend due to the term 10*x + 3, but the added random integers introduce significant vertical scatter.
- This spread represents the variability or randomness in the data.

## plt.scatter on Pandas dataframes üìà

```python
df = pd.read_csv('batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='o') #plt.scatter(x,y) -> x = df['avg'] -> y = df['strike_rate']
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

### Relationship Analysis:
- The scatter plot visually depicts the relationship between batting average (avg) and strike rate (strike_rate) for the top 50 batsmen.
- Each point on the plot represents a batsman, with their average on the x-axis and their strike rate on the y-axis.
### Insight Gathering:
- Tightly clustered points: This suggests that top batsmen with high averages also tend to have consistently high strike rates. In other words, they consistently score well and quickly.
- Widely scattered points: This indicates that among top batsmen, there is more variation in how average (total runs scored per dismissal) relates to strike rate (number of runs scored per 100 balls faced). Some may have high averages but varying strike rates, or vice versa. It shows a less predictable relationship between these two factors among top performers.

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips['total_bill'],tips['tip'], color='red',marker='o')
plt.grid(True) #Help in better visualization
plt.xlabel('Total bill')
plt.ylabel('Tip get')
```

### Visualization of Tip Behavior:
- The scatter plot visualizes the relationship between the total bill amount (total_bill) and the tip amount (tip) given by customers.
- Each point on the plot represents a transaction or customer visit to the restaurant.
### Insights from the Plot:
- You can observe if there's any correlation between the total bill amount and the tip amount.
    - Positive Correlation: We might observe that as the total_bill increases, the tip also tends to increase. 
- Clustering of points may indicate common tipping behaviors for different bill amounts.
- Outliers or unusual patterns can also be identified, providing insights into customer behavior regarding tipping.
### Moreover:
- Wide Spread: If the points are widely spread along the y-axis for a given x-value, it indicates high variability in tipping behavior for a particular total bill amount. This can be due to personal tipping habits, service quality, or other factors.
- Tight Clusters: Tight clustering around certain x-values can indicate more consistent tipping behavior for specific bill amounts.
- **Outliers**:
    - High Tips: Points significantly above the main cluster might indicate unusually high tips. These outliers could be due to exceptional service or customer generosity.
    - Low Tips: Points significantly below the main cluster might indicate unusually low tips, possibly due to poor service or other negative factors.

## Size of marker üìà

```python
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20) # Sets the size of each marker based on the party size.
```

### Marker Size Represents Party Size:
- The size of each marker is proportional to the party size (number of people at the table). Larger markers indicate larger groups, and smaller markers indicate smaller groups.
- This additional visual element can help identify if there are any patterns or trends related to party size.

# Bar charts üìä
## Bar charts are one of the most common and versatile types of data visualization. They are used to display and compare the frequency, count, or other measures (such as mean, median, or sum) for different categories or groups.
## Detailed Description
### A bar chart typically consists of:
- Bars: Rectangles that represent the value of each category. The length or height of the bar is proportional to the value it represents.
- Axes: Two axes (usually horizontal and vertical). The horizontal axis (x-axis) displays the categories, and the vertical axis (y-axis) shows the values.
- Labels: Descriptive labels for the axes and bars to provide context and meaning to the data.
### Bar charts can be:
- Vertical (Column) Bar Charts: Bars are vertical, with the height representing the value.
- Horizontal Bar Charts: Bars are horizontal, with the length representing the value.
### Bar charts are suitable for:
- Comparing discrete categories: Showing differences among distinct categories.
- Displaying changes over time: Although line charts are usually better for continuous data over time, bar charts can be used for discrete time periods (e.g., sales per quarter).

## Simple bar chart üìä
